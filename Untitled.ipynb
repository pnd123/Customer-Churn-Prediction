```python
# Sample: Predicting a number using linear regression (supervised ML)
from sklearn.linear_model import LinearRegression
import numpy as np

# Example data: hours studied vs test score
X = np.array([[1], [2], [3], [4], [5]])   # hours studied
y = np.array([50, 55, 65, 70, 80])        # test scores

model = LinearRegression()
model.fit(X, y)
pred = model.predict([[6]])   # Predict score for 6 hours studied
print(pred)
```
